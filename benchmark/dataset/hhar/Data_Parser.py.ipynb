```python
import pandas as pd
```

```python
watch_acc_df = pd.read_csv("/Users/mtang/Documents/HHAR/Watch_accelerometer.csv")
```

```python
# watch_acc_df.head()
watch_acc_df = watch_acc_df.rename(columns={'x':'watch_acc_x', 'y': 'watch_acc_y', 'z' : 'watch_acc_z'})
watch_acc_df.head()
```

```python
watch_gyro_df = pd.read_csv("/Users/mtang/Documents/HHAR/Watch_gyroscope.csv")
```

```python
watch_gyro_df = watch_gyro_df.rename(columns={'x':'watch_gyro_x', 'y': 'watch_gyro_y', 'z' : 'watch_gyro_z'})
watch_gyro_df.head()
```

```python
phone_acc_df = pd.read_csv("/Users/mtang/Documents/HHAR/Phones_accelerometer.csv")
```

```python
phone_acc_df = phone_acc_df.rename(columns={'x':'phone_acc_x', 'y': 'phone_acc_y', 'z' : 'phone_acc_z'})
phone_acc_df.head()
```

```python
phone_gyro_df = pd.read_csv("/Users/mtang/Documents/HHAR/Phones_gyroscope.csv")
```

```python
phone_gyro_df = phone_gyro_df.rename(columns={'x':'phone_gyro_x', 'y': 'phone_gyro_y', 'z' : 'phone_gyro_z'})
phone_gyro_df.head()
```

```python
watch_df = watch_acc_df.merge(watch_gyro_df, how='outer')
```

```python
watch_df = watch_df.sort_values(by='Arrival_Time')
watch_df.head(20)
```

```python
phone_df = phone_acc_df.merge(phone_gyro_df, how='outer')
```

```python
phone_df = phone_df.sort_values(by='Arrival_Time')
phone_df.head(20)
```

```python
whole_df = pd.concat([phone_df, watch_df], ignore_index=True, axis=1)
```

```python
whole_df = whole_df.sort_values(by='Arrival_Time')
whole_df.head(100)
```
