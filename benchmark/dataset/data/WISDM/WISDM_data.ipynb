```python
from scipy.io import arff
import pandas as pd
```

```python
data, meta = arff.loadarff('/Users/mtang/Documents/FedScale/benchmark/dataset/data/WISDM/WISDM_ar_v1.1_transformed.arff')
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df["user"] = df["user"].apply(lambda x: x.decode("utf-8"))
df["class"] = df["class"].apply(lambda x: x.decode("utf-8"))
```

```python
df = df.reset_index(drop=True)
df = df.drop(['UNIQUE_ID'], axis=1)
```

```python
df = df.rename(columns={"user": "client_Id", "class": "label_name"})
df = df.fillna(0)
# df["label_Id"] = df["label_name"].apply(lambda x: pd.factorize(x)[0])
df['label_Id'] = pd.factorize(df['label_name'])[0]
df.head()
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2)
```
