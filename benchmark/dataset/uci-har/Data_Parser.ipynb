```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
# get feature names from the file features.txt
features = list()
with open('/Users/mtang/Documents/UCI HAR Dataset/features.txt') as f:
    features = [line.split()[1] for line in f.readlines()]

print('No of Features: {}'.format(len(features)))
```

```python
seen = set()
uniq_features = []
for idx, x in enumerate(features):
    if x not in seen:
        uniq_features.append(x)
        seen.add(x)
    elif x + 'n' not in seen:
        uniq_features.append(x + 'n')
        seen.add(x + 'n')
    else:
        uniq_features.append(x + 'nn')
        seen.add(x + 'nn')
len(uniq_features)
```

```python
# get the data from txt files to pandas dataffame
X_train = pd.read_csv('/Users/mtang/Documents/UCI HAR Dataset/train/X_train.txt', delim_whitespace=True, header=None, names=uniq_features)

# # add subject column to the dataframe
X_train['subject'] = pd.read_csv('/Users/mtang/Documents/UCI HAR Dataset/train/subject_train.txt', header=None, squeeze=True)

y_train = pd.read_csv('/Users/mtang/Documents/UCI HAR Dataset/train/y_train.txt', names=['Activity'], squeeze=True)
y_train_labels = y_train.map({1: 'WALKING', 2:'WALKING_UPSTAIRS',3:'WALKING_DOWNSTAIRS',\
                       4:'SITTING', 5:'STANDING',6:'LAYING'})

# # put all columns in a single dataframe
train = X_train
train['Activity'] = y_train
train['ActivityName'] = y_train_labels
train.head()
```

```python
# get the data from txt files to pandas dataffame
X_test = pd.read_csv('/Users/mtang/Documents/UCI HAR Dataset/test/X_test.txt', delim_whitespace=True, header=None, names=uniq_features)

# add subject column to the dataframe
X_test['subject'] = pd.read_csv('/Users/mtang/Documents/UCI HAR Dataset/test/subject_test.txt', header=None, squeeze=True)

# get y labels from the txt file
y_test = pd.read_csv('/Users/mtang/Documents/UCI HAR Dataset/test/y_test.txt', names=['Activity'], squeeze=True)
y_test_labels = y_test.map({1: 'WALKING', 2:'WALKING_UPSTAIRS',3:'WALKING_DOWNSTAIRS',\
                       4:'SITTING', 5:'STANDING',6:'LAYING'})


# put all columns in a single dataframe
test = X_test
test['Activity'] = y_test
test['ActivityName'] = y_test_labels
test.head()
```

```python
columns = train.columns

# Removing '()' from column names
columns = columns.str.replace('[()]','')
columns = columns.str.replace('[-]', '')
columns = columns.str.replace('[,]','')

train.columns = columns
test.columns = columns

test.columns
```

```python
train = train.drop(columns=('ActivityName'))
```

```python
first_column = train.pop('subject')
train.insert(0, 'client_Id', first_column)
train.head()
```

```python
last_column = train.pop('Activity')
train.insert(len(train.columns), 'label_Id', last_column)
train.head()
```

```python
test = test.drop(columns=('ActivityName'))
first_column = test.pop('subject')
test.insert(0, 'client_Id', first_column)
last_column = test.pop('Activity')
test.insert(len(test.columns), 'label_Id', last_column)
test.head()
```

```python
whole_df = pd.concat([train, test])
whole_df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(whole_df, test_size=0.2, random_state=42)
```

```python
train.to_csv('/Users/mtang/Documents/FedScale/benchmark/dataset/data/UCI-HAR/client_data_mapping/train.csv', index=False)
test.to_csv('/Users/mtang/Documents/FedScale/benchmark/dataset/data/UCI-HAR/client_data_mapping/test.csv', index=False)
```

```python
import random
```

```python
modalities = []
for feature in uniq_features:
    modality = feature.split("-")[0]
    modalities.append(modality)
```

```python
group_list = []
client_map_modality = {}
group_df = whole_df.groupby('client_Id')
for name, group in group_df:
    modality = random.choice(modalities)
    client_map_modality[name] = modality
    for column in group.columns:
        if column.startswith(modality):
            group[column] = 0
    group_list.append(group)
```

```python
missing_group_df = pd.concat(group_list)
```

```python
missing_group_df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(missing_group_df, test_size=0.2, random_state=42)
```

```python
train.to_csv('/Users/mtang/Documents/FedScale/benchmark/dataset/data/UCI-HAR_MISSING/client_data_mapping/train.csv', index=False)
test.to_csv('/Users/mtang/Documents/FedScale/benchmark/dataset/data/UCI-HAR_MISSING/client_data_mapping/test.csv', index=False)
```
